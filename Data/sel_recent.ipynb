```python
import pandas as pd
```

```python
df_1= pd.read_csv("Adolescent_birth_rate.csv")
df_2= pd.read_csv("Informed_decision.csv")
```

```python
def sel_recent_year(df):
    df_c= df.copy()
    df_c= df_c.groupby("Geographic Area Code").max("Year")
    return df_c
```

```python
def filter_out(df1, df2):
    merged_df = df1.merge(df2, on='Geographic Area Code', suffixes=('', '_max'))
    filtered_df = merged_df[merged_df['Year'] == merged_df['Year_max']]
    final_df = filtered_df.drop('Year_max', axis=1)
    return final_df
```

```python
df_1= filter_out(df1= df_1, df2= sel_recent_year(df_1))
display(df_1)
df_1.to_csv("Adolescent_birth_rate.csv")
```

```python
df_2= filter_out(df1= df_2, df2= sel_recent_year(df_2))
display(df_2)
df_2.to_csv("Informed_decision.csv")
```
